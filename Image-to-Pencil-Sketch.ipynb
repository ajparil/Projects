## Project Name: Converting an Image to a Pencil Sketch 

This project aims to transform digital images into realistic pencil sketches using Python and image processing techniques. The goal is to offer a user-friendly solution for enhancing the artistic appeal of photos through processes such as grayscale conversion, inversion, and Gaussian blur

### Import the Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
print("Necessary packages included successfully!")
```

### Read the Image

```python
image=cv2.imread("cat.jfif")
image = cv2.resize(image,(500,300))
image
```

```python
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

### Convert the Original Image to Grayscale

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
gray_image
```

```python
plt.imshow(gray_image)
```

### Inverting the Image

```python
inverted_image = cv2.bitwise_not(gray_image)
inverted_image
plt.imshow(inverted_image)
```

```python
img_smoothing = cv2.GaussianBlur(inverted_image, (21, 21),sigmaX=0, sigmaY=0)
img_smoothing
plt.imshow(img_smoothing)
```

```python
def dodgeV2(x, y):
    return cv2.divide(x, 255 - y, scale=256.0)
```

```python
pencil_sketch = dodgeV2(gray_image, img_smoothing)
plt.imshow(pencil_sketch)
```

```python
display(Image.fromarray(img_rgb))
display(Image.fromarray(pencil_sketch))
```

```python
cv2.imshow("Sketch", pencil_sketch)
cv2.waitKey(0)
```

The pencil sketch generation project skillfully employed image processing techniques such as grayscale conversion, smoothing, edge detection, and binary thresholding to transform digital images into realistic pencil sketches which successfully transform a digital image of a cat into a realistic pencil sketch. This project provides a clear and effective approach to creating visually appealing sketches, offering both insight into the image processing pipeline and a practical implementation for users interested in generating artistic effects from their digital images.
